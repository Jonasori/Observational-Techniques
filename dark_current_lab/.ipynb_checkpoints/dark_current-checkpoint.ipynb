**Measuring the Read Noise in ADU**

*The read noise of the CCD causes the ADU recorded for each pixel to differ from the ADU value that should be produced given the charge deposited in the pixel. Using the bias frames, we can measure the read noise by calculating the mean and standard deviation of the pixel values (after removing hot and dead pixels). The standard deviation is often described as the RMS (or rms) since it is the square root of the mean of the squared deviations. We can use Python to calculate these statistics and plot a histogram of the pixel values in our bias frames.*

We begin by defining a function that gives us image stats, a histogram of count distributions, and the image itself:

```python
def get_img_stats(img, plow=1861, phi=2000, cut_by_std=True):
    nx, ny = img.shape
    # Change the array to 1D for histogram plotting.
    imgh = np.reshape(img, nx*ny)

    # Cut out bad data (dead/hot pixels)
    if cut_by_std is True:
        im_mean, im_std = np.mean(imgh), np.std(imgh)
        print "Drawing from data in 5-sigma range around:"
        print "Mean, std: ", im_mean, im_std
        # Take the most restrictive set possible? I guess?
        plow = max(plow, im_mean - 5 * im_std)
        phi = min(phi, im_mean + 5 * im_std)
    
    q = np.where((imgh >= plow) & (imgh <= phi))
    imghcut = imgh[q]

    print 'Image minimum = ', np.nanmin(imghcut)
    print 'Image maximum = ', np.nanmax(imghcut)
    print 'Image mean = ', np.mean(imghcut)
    print 'Image standard deviation = ', np.std(imghcut)

    fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 3))
    ax1.imshow(img, cmap=colmap)
    ax2.hist(imghcut, bins=100, histtype='stepfilled')
    
    ax1.set_xticks([])
    ax1.set_yticks([])
    nc = max(plt.hist(imghcut, bins=100, histtype='stepfilled')[0])
    ax2.set_xticks([min(imghcut), max(imghcut) - (max(imghcut) - min(imghcut))/2, max(imghcut)])
    ax2.set_yticks([0, nc/2, nc])
    ax2.set_xlabel('Count')
    ax2.set_ylabel('Frequency')
    plt.show()
```

We may now run it on one of our bias images.

```python
img = fits.open(basepath + "bias/bias_-5deg.fits")[0].data.astype('int16')
get_img_stats(img, plow=0, phi=5000, cut_by_std=True)
```

As anticipated, our count frequencies are piled in a Guassian distribution around some central value (the bias) and our image doesn't look like much of anything, because all the pixels have basically the same count since the deviation in our count distribution is small relative to the number of counts.

```python
print "Beginning calculations...\n\n"

# Read in the bias file
bias = fits.open(basepath + 'darks/dark_0sec_0deg.fits',)[0].data.astype('int16')

# list of dark files
darkfile = [basepath + 'darks/dark_1sec_0deg.fits',
            basepath + 'darks/dark_5sec_0deg.fits',
            basepath + 'darks/dark_20sec_0deg.fits',
            basepath + 'darks/dark_60sec_0deg.fits',
            basepath + 'darks/dark_180sec_0deg.fits',
            basepath + 'darks/dark_600sec_0deg.fits',]
# array of corresponding times
time = np.array([1, 5, 20, 60, 180, 600])

# arrays to hold results
c_mean = 0.0*time
c_median = 0.0*time
c_rms = 0.0*time

# Process the files
plt.close()
fig, ax = plt.subplots(nrows=len(darkfile), ncols=2, figsize=(10, 12))
for i in range(len(darkfile)):
    # Read in the file
    img = fits.open(darkfile[i])[0].data.astype('int16')

    diff = img - bias
    nx, ny = diff.shape # find the size of the array
    diffh = np.reshape(diff, nx*ny) # change the shape to be 1d
    s = np.sort(diffh)

    f = 0.03 # ignore first and last fraction f of points
    vmin = s[int(f*len(s))]
    vmax = s[int((1-f)*len(s))]
    vmax = 100
    # Select only values within ranges
    q = np.where((diffh >= vmin) & (diffh <= vmax))
    diffhcut = diffh[q]


    # Find and save statistics
    # There's gotta be a better way to do this.
    nc = plt.hist(diffh, bins=int(vmax-vmin), range=[vmin,vmax], histtype='stepfilled', color='k')[0]
    c_mean[i]   = np.mean(diffhcut)
    c_median[i] = np.median(diffhcut)
    c_rms[i]    = np.std(diffhcut)

    print 'Mean, median, RMS = ', c_mean[i], c_median[i], c_rms[i]


    ax[i][0].imshow(diff, cmap=colmap, vmin=vmin, vmax=vmax)
    ax[i][1].hist(diffh, bins=vmax-vmin, range=[vmin,vmax],
                  histtype='stepfilled', color='k')

    # Plot vertical lines on the histogram to tell us things.
    ax[i][1].plot([c_mean[i], c_mean[i]], [min(nc), max(nc)], '-g', label='Mean')
    ax[i][1].plot([c_median[i], c_median[i]], [min(nc), max(nc)], '--r', label='Median')
    ax[i][1].legend()

    # Make the plots look a little nicer.
    ax[i][0].set_xticks([])
    ax[i][0].set_yticks([])
    ax[i][1].set_xticks([min(diffhcut),  0, max(diffhcut)])
    ax[i][1].set_yticks([min(nc), max(nc)])

    ax[i][1].set_xlabel("Photon Count")
    ax[i][1].set_ylabel("Frequency")
plt.show()


# Do a linear fit to the data
slope_mean, intercept_mean, r_value_mean, p_value_mean, std_err_mean = stats.linregress(time, c_mean)
slope_median, intercept_median, r_value_median, p_value_median, std_err_median = stats.linregress(time, c_median)
# print('Slope = ', slope)
# print('Intercept = ', intercept)
# print('Correlation coefficient r =', r_value)

# Plot the fit

plt.plot(time, c_mean, 'd', color='green')
plt.plot(time, c_median, 'd', color='red')
plt.plot(time, intercept_mean + time*slope_mean, '--g', label='Mean')
plt.plot(time, intercept_median + time*slope_median, '--r', label='Median')    

plt.xlabel('Time (s)')
plt.ylabel('Counts')
plt.legend()
plt.annotate("R-value for medians: " + str(round(r_value_median, 4)), (0.1*max(time), 0.9*max(c_mean)))
plt.annotate("R-value for means: " + str(round(r_value_mean, 4)), (0.1*max(time), 0.8*max(c_mean)))

plt.annotate("Slope value for medians: " + str(round(slope_median, 4)), (0.5*max(time), 0.35*max(c_mean)))
plt.annotate("Slope value for means: " + str(round(slope_mean, 4)), (0.5*max(time), 0.25*max(c_mean)))

plt.show()
```

*Make a plot of counts (mean or median) versus temperature. Then program in the conversions to make a plot of dark current in terms of electrons/pixel/second (on the y-axis with a log scale) versus temperature (on the x-axis with a linear scale).*

```python
dark_current = c_mean * DCelectrons_pixel_second
plt.semilogy(temps, dark_current)
plt.semilogy(temps, dark_current, marker='o')

plt.xlabel('Temp (Celcius)')
plt.ylabel('Dark Current Electrons/pixel/sec')
plt.show()
```

This is kinda garbage, since it's implying that the dark current actually decreases as temperature increases, which we know to be wrong. It is probably showing this because we are too cold and so the CCD hasn't actually entered the linear (well, really exponential) temp/dark current regime yet.

*Now we want to fit an exponential function to the data, i.e. dark current = A*exp(B*temperature). You can do this using the scipy.stats.linregress (**or scipy.optimize.curve_fit**) function. Plot your fit on your data, then save the plot.*

```python
# Do a linear fit to the data
def predict_count(t, a, b, c):
    return a * np.exp(-b * t) + c


popt_median, pcov_median = curve_fit(predict_count, temps, c_median)
popt_mean, pcov_mean = curve_fit(predict_count, temps, c_mean)

# Plot the fit
plt.plot(temps, c_mean, 'd', color='green')
plt.plot(temps, c_median, 'd', color='red')
plt.plot(temps, predict_count(temps, *popt_mean), '--g', label='Fit Means')
plt.plot(temps, predict_count(temps, *popt_median), '--r', label='Fit Medians')

plt.xlabel('Time (s)')
plt.ylabel('Counts')
plt.legend()
plt.show()

a_med, b_med, c_med = popt_median
print "Median Exponential Fits: y = " + str(round(a_med, 3)) + " * exp(-" + str(round(b_med, 3)) + " t) + " + str(round(c_med, 3))

a_avg, b_avg, c_avg = popt_mean
print "Mean Exponential Fits:   y = " + str(round(a_avg, 3)) + " * exp(-" + str(round(b_avg, 3)) + " t) + " + str(round(c_avg, 3))
```

Since we're in the bad regime as discussed above, the fit is naturally kinda meaningless.
